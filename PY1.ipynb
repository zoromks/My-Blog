```python
def fact(a):
    if a==0:
        return 1
    return a*fact(a-1)

fact(8)
```

```python
class prnt:
    def __init__(self):
        self.s=''
    def to_get(self):
        self.s=input()
    def to_prt(self):
        print(self.s)

a=prnt()
a.to_get()
a.to_prt()
```

```python
import math
c=50
h=30
d=[x for x in input().split(',')]
val=[]
for i in d:
    val.append(str(round(math.sqrt((2 * c * int(i))/h))))
print(','.join(val))
```
